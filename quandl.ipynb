```python
import Quandl
import pandas as pd
from ggplot import *
% matplotlib inline
```

```python
index = Quandl.get('ZILL/N00146_C')
sfl = Quandl.get('ZILL/N00146_SFL')
index['Year'] = index.index
sfl['Year'] = sfl.index
```

```python
index = Quandl.get('ZILL/Z92122_C')
sfl = Quandl.get('ZILL/Z92122_SFL')
index['Year'] = index.index
sfl['Year'] = sfl.index
```

```python
ggplot(aes(x='Year', y = 'Value'), data = index)\
+ geom_line()\
+ ggtitle('UTC Condo Index')
```

```python
ggplot(aes(x='Year', y = 'Value'), data = sfl)\
+ geom_line()\
+ ggtitle('UTC Percent Sold For Loss')
```
