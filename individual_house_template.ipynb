```python
import pandas as pd
import numpy as np
import Quandl
from ggplot import *
% matplotlib inline
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999

import cost_helper
```

# Realistic Estimates

```python
inputs = cost_helper.calculate_inputs()
cost = cost_helper.calculate_cost(inputs)
```

```python
total_plot = pd.melt(cost, id_vars = ['Year'], value_vars = ['Opportunity_Cost','House_Net_Value','Rent_Net_Value'])
total_plot.set_index([range(len(total_plot))], inplace = True)
total_plot['value'] = np.round(total_plot['value'])
ggplot(aes(x = 'Year', y = 'value/1000', color = 'variable'), data = total_plot)\
+ geom_line()\
+ theme_seaborn(context = 'poster')\
+ ylab('Value (1k USD)')
#+ xlim(0,10)
```
