### This cell you edit once and is assumed to be independent of the actual house you are looking at

```python
# Static Inputs

inputs['yearly_house_appreciation'] = .02 # percent appreciation of house value

# Taxes
inputs['marginal_tax_rate'] = .25
#inputs['expected_deduction'] = 6400 # expected standard deduction for current year, only relevant if you are taking standard deduction and then going to itemize after buying house
#inputs['expected_deduction_yearly_increase'] = 100 # expected increase in standard deduction each year,
inputs['expected_deduction'] = 0 # expected standard deduction for current year, set to zero if you always itemize deduction
inputs['expected_deduction_yearly_increase'] = 0 # expected increase in standard deduction each year, set to zero if always itemize deductions
inputs['property_tax'] = .01 # percent of house value payed as yearly property tax
inputs['long_term_capital_gains'] = .15

#Extra Costs
inputs['yearly_maintenance'] = .003 # percent of house value set aside yearly for repairs and maintenance
inputs['hoa_yearly_increase'] = .02 # yearly percent increase in hoa fee
inputs['homeowner_yearly_increase'] = .01 # yearly percent increase in insurance
inputs['closing_buy'] = .06 # percent of house value for closing cost when buying
inputs['closing_sell'] = .06 # percent of house value for closing cost when selling

#Rental Income
inputs['rent_yearly_increase'] = .02 # yearly percent increase in rent
inputs['percent_vacancy'] = 1/12.0 # percent of the year with vacancy
inputs['rent_percent'] = 0.0 #percent of monthly rent as income, 0 means not renting, .5 means renting out half your house value, functionality for rental property not included yet

# Oportunity costs
inputs['investment_return'] = .07 #average yearly return
inputs['opportunity_rent'] = inputs['monthly_rent'] * (1 - inputs['rent_percent']) # amount you would pay to rent instead
```

### By default, I will pull relevant data from zillow:

```python
#replace some user inputs with zillow provided data
use_zillow = True # set to false to ignore this cell
address = '9082 Flanders Dr'.replace(' ','+')
city = 'San Diego'
state = 'CA'
zip_code = '92126'

if use_zillow == True:
    zillow_data = zillow_helper.get_zillow_data(address,city,state,zip_code)
    if inputs['time'] == 30:
        inputs['interest'] = zillow_data['thirty_year_rate']/100.0 # APY interest rate
    else:
        inputs['interest'] = zillow_data['fifteen_year_rate']/100.0 # APY interest rate
    inputs['initial_value'] = zillow_data['zestimate']
    inputs['monthly_rent'] = zillow_data['rent_zestimate']
    inputs['opportunity_rent'] = inputs['monthly_rent'] * (1 - inputs['rent_percent']) # amount you would pay to rent instead

    
        
```

```python
house = house_master.House(inputs)
house.calculate_cost(house.inputs)
plot_data = house.Plot_Data
```

### Plot:

Key Assumptions:  
1. No capital gains when you sell your house (google for details)
2. House Value, HOA, Maintenance, Insurance, and Rent all increase/decrease by the same amount each year. Not a problem for long-term if good estimates.
3. You stick with one rental strategy.
4. Fixed Rate
5. No PMI
6. All gains on stocks are long term capital gains

```python
total_plot = pd.melt(plot_data, id_vars = ['Year'], value_vars = ['Opportunity_Cost','House_Net_Value','Rent_Net_Value'])
ggplot(aes(x = 'Year', y = 'value/1000', color = 'variable'), data = total_plot)\
+ geom_line()\
+ theme_seaborn(context = 'poster')\
+ ylab('Value (1k USD)')
```

**House_Net_Value**: This is the amount of money in your pocket after you live there for x years and sell it on year x.  
**Rent_Net_Value**: This is the amount of money you have after x years if you rent instead and invest your initial cost (downpayment and closing) and any extra money you save by renting.
